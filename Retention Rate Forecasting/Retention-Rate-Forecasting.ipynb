## Transform & explore data

```python
#importing the package
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing the data
data = pd.read_csv('Employee Atrrition.csv')
data.head()
```

```python
data.describe()
```

```python
#Turn the attrition into numbers 1 and 0 
data["Attrition"].replace({"Yes": 1, "No": 0}, inplace=True)
data["BusinessTravel"].replace({"Non-Travel": 1, "Travel_Rarely": 2, "Travel_Frequently": 3}, inplace=True)
data["OverTime"].replace({"Yes": 1, "No": 0}, inplace=True)
data["Gender"].replace({"Female": 1, "Male": 0}, inplace=True)
```

```python
data.head()
```

```python
#Instead of using only age feature, I also want to see if agegroup might impact on attrition or not
AgeGroup = []
for i,k in enumerate (data['Age']):
    if k<25:
        AgeGroup.append(1) #GenZ
    elif 24<k<34:
        AgeGroup.append(2) #Millenials
    elif 34<k<55:
        AgeGroup.append(3) #Gen X
    else:
        AgeGroup.append(4) #Baby boomers
```

```python
data['AgeGroup'] = AgeGroup
```

```python
#Using one-hot coding to transform categorical features
data = pd.get_dummies(data, columns = ['Department'])
data = pd.get_dummies(data, columns = ['JobRole'])
data = pd.get_dummies(data, columns = ['MaritalStatus'])
data = pd.get_dummies(data, columns = ['EducationField'])
```

```python
data.head()
```

```python
#Now I want to see the unique values of each columns and count => in order to see some disastisfied columns 
for column in data.columns :
    if data[column].dtype  == object :
        print(str(column) + ' : ' + str(data[column].unique()))
        print(data[column].value_counts())
        print('_____________________________')
```

```python
#So we can see in these line, Over18 just has 1 variable is 'Y' , so we need to drop this columns
#But I want to exclude other features that are not related to the variable 
data.corr()['Attrition'].sort_values()
# We have more 3 columns need to drop out 
data = data.drop(columns={'EmployeeCount','StandardHours','Over18','EmployeeNumber'})
data
```

```python
#Take a look at feature correlation to see which ones might impact the most on attrition
plt.figure(figsize=(30,12))
sns.heatmap(data.corr(), annot=True, fmt ='.0%')
plt.show()
#agegroup, job level, total working years, monthly income, years with current managers, years with current role
#'Attrition','BusinessTravel', 'JobLevel', 'MonthlyIncome', 'TotalWorkingYears','YearsWithCurrManager','YearsInCurrentRole','AgeGroup', 'MaritalStatus', 'JobRole'
```

## Logistics Regression

```python
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
#selecting biggest and smallest correlation plots, make sure each has <50% correlation with other features
data1 = data[['Attrition', 'OverTime','BusinessTravel', 'JobLevel', 'YearsWithCurrManager','AgeGroup', 'Department_Research & Development','JobRole_Sales Representative', 'MaritalStatus_Single']]
data1.head()
              
```

```python
Y = data1.Attrition==1
features= [x for x in data1.columns if x!='Attrition']
X = pd.get_dummies(data1[features])
X_train, X_test, Y_train, Y_test = train_test_split( X ,  Y , test_size= 0.2, random_state = 40)
```

```python
#import your model
model=LogisticRegression(max_iter = 20000)

#fitting and predict your model
model.fit(X_train, Y_train)
Y_pre = model.predict(X_test)
```

```python
result = pd.DataFrame({'Y_test':Y_test, 'prediction':model.predict(X_test)}).head(20)
result
```

```python
confusion_matrix = pd.crosstab(Y_test, Y_pre, rownames=['Actual'], colnames = ['Predict'])
confusion_matrix
```

```python
metrics.accuracy_score(Y_test, Y_pre)
print ('Accuracy: ', metrics.accuracy_score(Y_test, Y_pre))
plt.show()
```
